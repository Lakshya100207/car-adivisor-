```python
import pandas as pd  # For data handling
import numpy as np   # For calculations
import os            # For file operations
from pathlib import Path  # For path handling

print("âœ… Libraries imported successfully!")
```

```python
# List all CSV files in datasets folder
folder_path = "../datasets/"  # Go up one level, then into datasets

files = []
for file in os.listdir(folder_path):
    if file.endswith(".csv"):
        files.append(file)
        print(f"ðŸ“„ Found: {file}")

print(f"\nTotal CSV files found: {len(files)}")
```

```python
# Load just ONE file first to understand the data
file_name = files[0]  # Takes the first file
file_path = f"../datasets/{file_name}"

# Read the CSV
df1 = pd.read_csv(file_path)

# Show basic info
print(f"ðŸ“Š File: {file_name}")
print(f"Rows: {df1.shape[0]}, Columns: {df1.shape[1]}")
print("\nFirst 5 rows:")
display(df1.head())  # Shows the first 5 rows
print("\nColumn names:")
print(list(df1.columns))
```

```python
# Create dictionary to store all datasets
all_data = {}

# Load each CSV file
for file in files:
    try:
        file_path = f"../datasets/{file}"
        df = pd.read_csv(file_path)
        all_data[file] = df
        print(f"âœ… Loaded {file}: {df.shape[0]} rows x {df.shape[1]} columns")
    except Exception as e:
        print(f"Error loading {file}: {e}")

print(f"\nTotal datasets loaded: {len(all_data)}")
```

```python
# Display info about each dataset
for name, df in all_data.items():
    print(f"\n{'='*50}")
    print(f"Dataset: {name}")
    print(f"{'='*50}")
    print(f"Shape: {df.shape}")
    print("Columns:", list(df.columns))
    print("\nFirst 3 rows:")
    display(df.head(3))
```

```python
# Method 1: Simple combine (stack on top of each other)
combined_data = pd.DataFrame()  # Create empty dataframe

for name, df in all_data.items():
    # Add a column to remember source
    df['source_file'] = name
    # Add to combined data
    combined_data = pd.concat([combined_data, df], ignore_index=True)

print(f"âœ… Combined all datasets!")
print(f"Total rows: {combined_data.shape[0]}")
print(f"Total columns: {combined_data.shape[1]}")
print("\nFirst 5 rows:")
display(combined_data.head())
```

```python
print("ðŸ” UNDERSTANDING THE COLUMN PROBLEM")
print("="*60)

# Count how many non-empty values in each column
column_stats = []
for col in combined_data.columns:
    non_null = combined_data[col].notna().sum()
    if non_null > 0:  # Only show columns with some data
        column_stats.append((col, non_null))

# Sort by most populated columns
column_stats.sort(key=lambda x: x[1], reverse=True)

print(f"\nTop 20 columns with most data:")
for i, (col, count) in enumerate(column_stats[:20], 1):
    percentage = (count / len(combined_data)) * 100
    print(f"{i:2}. {col:30} â†’ {count:5} rows ({percentage:.1f}%)")

print(f"\nðŸ“Š Total columns with data: {len(column_stats)}")
print(f"ðŸ“Š Empty columns: {346 - len(column_stats)}")
```

```python
print("IDENTIFYING DATASET TYPES")


# Let's check each dataset individually
for name, df in all_data.items():
    print(f"\n {name}:")
    print(f"  Rows: {df.shape[0]}, Columns: {df.shape[1]}")
    
    # Check for key columns to determine type
    has_new_car_cols = 'Brand' in df.columns and 'Model' in df.columns
    has_used_car_cols = 'Car_ID' in df.columns or 'Owner_Type' in df.columns
    
    if has_new_car_cols:
        print(" Type: NEW CAR SPECIFICATIONS")
        new_car_cols = [col for col in ['Brand', 'Model', 'Fuel_Type', 'Seating_Capacity', 'Body_Type'] if col in df.columns]
        print(f"  Key columns: {new_car_cols}")
        
    elif has_used_car_cols:
        print("  Type: USED CAR LISTINGS")
        used_car_cols = [col for col in ['Car_ID', 'Owner_Type', 'Previous_Accidents', 'Service_Cost'] if col in df.columns]
        print(f"  Key columns: {used_car_cols}")
        
    else:
        print("  Type: UNKNOWN/MIXED")
    
    # Show first few column names
    print(f"  First 5 columns: {list(df.columns)[:5]}")
```

```python
print("ðŸŽ¯ LOADING KEY NEW CAR DATASETS")
print("="*60)

# Load the 3 main datasets
key_datasets = {
    'dataset2': all_data['dataset2.csv'],
    'dataset6': all_data['dataset6.csv'], 
    'dataset7': all_data['dataset7.csv']
}

# Show detailed info for each
for name, df in key_datasets.items():
    print(f"\n{'='*50}")
    print(f"ðŸ“Š {name}: {df.shape[0]} rows, {df.shape[1]} columns")
    print(f"{'='*50}")
    
    print("FIRST 3 ROWS:")
    display(df.head(3))
    
    print("\nCOLUMNS AVAILABLE:")
    for i, col in enumerate(df.columns, 1):
        # Show first non-null value as example
        sample_val = df[col].dropna().iloc[0] if not df[col].dropna().empty else "EMPTY"
        print(f"{i:3}. {col:30} â†’ Sample: {str(sample_val)[:50]}...")
```

```python
print("ðŸ”§ CREATING SMART COLUMN MAPPING")
print("="*60)

# For dataset2 (already has good structure)
dataset2_mapping = {
    'Brand': 'brand',
    'Model': 'model',
    'Price': 'price',
    'Fuel_Type': 'fuel_type',
    'Engine_CC': 'engine_cc',
    'Transmission': 'transmission',
    'Mileage': 'mileage_kmpl',
    'Body_Type': 'body_type',
    'Seating_Capacity': 'seating_capacity',
    'Year': 'model_year'
}

# For dataset6 (has Make/Model/Variant structure)
dataset6_mapping = {
    'Make': 'brand',
    'Model': 'model',
    'Variant': 'variant',
    'Ex-Showroom_Price': 'price',
    'Fuel_Type': 'fuel_type',
    'Engine_Displacement': 'engine_cc',
    'Transmission_Type': 'transmission',
    'ARAI_Certified_Mileage': 'mileage_kmpl',
    'Body_Type': 'body_type',
    'Seating_Capacity': 'seating_capacity'
}

# For dataset7 (has Brand/Car/Variant structure)
dataset7_mapping = {
    'Brand': 'brand',
    'Car': 'model',
    'Variant': 'variant',
    'Price': 'price',
    'Fuel_Type': 'fuel_type',
    'Engine': 'engine_cc',
    'Transmission': 'transmission',
    'Mileage': 'mileage_kmpl',
    'Body_Type': 'body_type',
    'Seats': 'seating_capacity'
}

print("Dataset 2 mapping:")
for old, new in dataset2_mapping.items():
    if old in key_datasets['dataset2'].columns:
        print(f"  âœ“ '{old}' â†’ '{new}'")

print("\nDataset 6 mapping:")  
for old, new in dataset6_mapping.items():
    if old in key_datasets['dataset6'].columns:
        print(f"  âœ“ '{old}' â†’ '{new}'")

print("\nDataset 7 mapping:")
for old, new in dataset7_mapping.items():
    if old in key_datasets['dataset7'].columns:
        print(f"  âœ“ '{old}' â†’ '{new}'")
```

```python
print("ðŸ”„ STANDARDIZING AND MERGING 3 KEY DATASETS")
print("="*60)

# Function to apply mapping
def standardize_dataset(df, mapping, source_name):
    df_clean = df.copy()
    
    # Apply column renaming
    df_clean.rename(columns=mapping, inplace=True)
    
    # Add source
    df_clean['source_dataset'] = source_name
    
    # Keep only the columns we want
    desired_cols = [
        'brand', 'model', 'variant', 'price', 'fuel_type', 
        'engine_cc', 'transmission', 'mileage_kmpl', 'body_type',
        'seating_capacity', 'model_year', 'source_dataset'
    ]
    
    # Create result with only desired columns
    result = pd.DataFrame()
    for col in desired_cols:
        if col in df_clean.columns:
            result[col] = df_clean[col]
    
    return result

# Process each dataset
print("Processing dataset2...")
df2_clean = standardize_dataset(key_datasets['dataset2'], dataset2_mapping, 'dataset2')
print(f"  âœ“ Rows: {df2_clean.shape[0]}, Columns: {df2_clean.shape[1]}")

print("\nProcessing dataset6...")
df6_clean = standardize_dataset(key_datasets['dataset6'], dataset6_mapping, 'dataset6')
print(f"  âœ“ Rows: {df6_clean.shape[0]}, Columns: {df6_clean.shape[1]}")

print("\nProcessing dataset7...")
df7_clean = standardize_dataset(key_datasets['dataset7'], dataset7_mapping, 'dataset7')
print(f"  âœ“ Rows: {df7_clean.shape[0]}, Columns: {df7_clean.shape[1]}")

# Remove completely empty columns to avoid warning
df2_clean = df2_clean.dropna(axis=1, how='all')
df6_clean = df6_clean.dropna(axis=1, how='all')
df7_clean = df7_clean.dropna(axis=1, how='all')

print("\n" + "="*50)
print("MERGING ALL STANDARDIZED DATASETS...")
merged_new_cars = pd.concat([df2_clean, df6_clean, df7_clean], ignore_index=True)
print(f"âœ… MERGED SUCCESSFULLY!")
print(f"Total cars: {merged_new_cars.shape[0]}")
print(f"Total columns: {merged_new_cars.shape[1]}")

print("\nðŸ“Š DATA COVERAGE AFTER MERGE:")
for col in merged_new_cars.columns:
    if col != 'source_dataset':
        non_null = merged_new_cars[col].notna().sum()
        percentage = (non_null / len(merged_new_cars)) * 100
        print(f"  â€¢ {col:20} â†’ {non_null:6} rows ({percentage:5.1f}%)")
```

```python
print("ðŸ§ª QUICK CHECK - IS MERGE WORKING?")
print("="*50)

if 'merged_new_cars' in locals():
    print(f"âœ… Merge variable exists!")
    print(f"Rows: {merged_new_cars.shape[0]}")
    print(f"Columns: {merged_new_cars.shape[1]}")
    
    # Show column names
    print("\nðŸ“‹ Columns found:")
    for col in merged_new_cars.columns:
        print(f"  â€¢ {col}")
    
    # Show sample
    print("\nðŸ‘€ Sample data:")
    display(merged_new_cars.head(3))
else:
    print("âŒ Merge not completed. Re-run Step 3.")
```

```python
print("ðŸ§¹ CLEANING AND DEDUPLICATING")
print("="*60)

print(f"Before cleaning: {merged_new_cars.shape[0]} rows")

# Create a unique identifier for each car
def create_car_id(row):
    # Try to create ID from available info
    parts = []
    if pd.notna(row.get('brand')):
        parts.append(str(row['brand']).lower().replace(' ', '_'))
    if pd.notna(row.get('model')):
        parts.append(str(row['model']).lower().replace(' ', '_'))
    if pd.notna(row.get('variant')):
        parts.append(str(row['variant']).lower().replace(' ', '_')[:20])
    return '_'.join(parts) if parts else None

merged_new_cars['car_id'] = merged_new_cars.apply(create_car_id, axis=1)

# Remove rows without car_id
initial_count = len(merged_new_cars)
merged_new_cars = merged_new_cars[merged_new_cars['car_id'].notna()]
print(f"Removed rows without ID: {initial_count - len(merged_new_cars)}")

# Remove exact duplicates
merged_new_cars = merged_new_cars.drop_duplicates(subset=['car_id'], keep='first')
print(f"Removed duplicate car entries: {initial_count - len(merged_new_cars)}")

print(f"\nâœ… After cleaning: {merged_new_cars.shape[0]} unique cars")

# Show sample
print("\nðŸ“‹ SAMPLE OF CLEANED DATA:")
sample_cols = ['car_id', 'brand', 'model', 'variant', 'price', 'fuel_type']
display(merged_new_cars[sample_cols].head(10))
```

```python
print("ðŸ” HANDLING MISSING VALUES")
print("="*60)

# Show missing values before
print("Missing values per column (before):")
for col in merged_new_cars.columns:
    if col not in ['source_dataset', 'car_id']:
        missing = merged_new_cars[col].isna().sum()
        if missing > 0:
            percentage = (missing / len(merged_new_cars)) * 100
            print(f"  â€¢ {col:20} â†’ {missing:5} missing ({percentage:5.1f}%)")

print("\nðŸ”„ Smart filling for same brand+model...")

# First, let's handle the easy cases with simple forward fill within groups
for col in ['fuel_type', 'transmission', 'body_type', 'seating_capacity']:
    if col in merged_new_cars.columns:
        # Group by brand and model, then fill within each group
        merged_new_cars[col] = merged_new_cars.groupby(['brand', 'model'])[col].transform(
            lambda x: x.ffill().bfill() if not x.isna().all() else x
        )

print("\nðŸ“Š After simple group filling:")
for col in ['fuel_type', 'transmission', 'body_type', 'seating_capacity']:
    if col in merged_new_cars.columns:
        missing = merged_new_cars[col].isna().sum()
        percentage = (missing / len(merged_new_cars)) * 100
        print(f"  â€¢ {col:20} â†’ {missing:5} missing ({percentage:5.1f}%)")

print("\nðŸ”„ Filling remaining missing values with overall mode...")

# For remaining missing values, fill with the overall most common value
for col in ['fuel_type', 'transmission', 'body_type', 'seating_capacity']:
    if col in merged_new_cars.columns:
        # Get the most common value (mode)
        mode_value = merged_new_cars[col].mode()
        if not mode_value.empty:
            # Fill remaining NaN with the most common value
            merged_new_cars[col] = merged_new_cars[col].fillna(mode_value.iloc[0])
            print(f"  â€¢ {col:20} â†’ Filled with: {mode_value.iloc[0]}")

print("\nâœ… Missing values per column (after all filling):")
for col in merged_new_cars.columns:
    if col not in ['source_dataset', 'car_id']:
        missing = merged_new_cars[col].isna().sum()
        if missing > 0:
            percentage = (missing / len(merged_new_cars)) * 100
            print(f"  â€¢ {col:20} â†’ {missing:5} missing ({percentage:5.1f}%)")
        else:
            print(f"  â€¢ {col:20} â†’ COMPLETE! (0 missing)")
```

```python
print("ðŸ”§ ADDITIONAL DATA CLEANING")
print("="*60)

# 1. Drop model_year if it's mostly empty
if 'model_year' in merged_new_cars.columns:
    missing_percentage = (merged_new_cars['model_year'].isna().sum() / len(merged_new_cars)) * 100
    if missing_percentage > 90:
        print(f"âš ï¸ Dropping 'model_year' column ({missing_percentage:.1f}% missing)")
        merged_new_cars = merged_new_cars.drop(columns=['model_year'])
    else:
        # Fill with a default year
        merged_new_cars['model_year'] = merged_new_cars['model_year'].fillna(2023)

# 2. Ensure price is numeric
if 'price' in merged_new_cars.columns:
    print("\nðŸ’° Cleaning price column...")
    # Convert to numeric, coerce errors
    merged_new_cars['price'] = pd.to_numeric(merged_new_cars['price'], errors='coerce')
    print(f"  â€¢ Non-numeric price values found: {(merged_new_cars['price'].isna().sum())}")

# 3. Fill variant with model name if missing
if 'variant' in merged_new_cars.columns and 'model' in merged_new_cars.columns:
    merged_new_cars['variant'] = merged_new_cars.apply(
        lambda row: row['model'] if pd.isna(row['variant']) else row['variant'],
        axis=1
    )
    print(f"  â€¢ Variant column now complete!")

# 4. Standardize fuel_type values
if 'fuel_type' in merged_new_cars.columns:
    print("\nâ›½ Standardizing fuel types...")
    fuel_mapping = {
        'petrol': 'Petrol',
        'diesel': 'Diesel',
        'electric': 'Electric',
        'hybrid': 'Hybrid',
        'cng': 'CNG'
    }
    
    def clean_fuel_type(val):
        if pd.isna(val):
            return val
        val_lower = str(val).lower().strip()
        for key, clean_val in fuel_mapping.items():
            if key in val_lower:
                return clean_val
        return str(val).strip()
    
    merged_new_cars['fuel_type'] = merged_new_cars['fuel_type'].apply(clean_fuel_type)
    print(f"  â€¢ Unique fuel types: {merged_new_cars['fuel_type'].nunique()}")

print("\nâœ… Additional cleaning complete!")
```
